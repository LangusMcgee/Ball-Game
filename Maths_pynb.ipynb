```python
import matplotlib.pyplot as plt
import numpy as np

def drawline(x1,y1,x2,y2):

    #work out differences.
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    #decision parameter
    dp = dx - dy

    # direction of line
    sx = 1 if x1 < x2 else -1
    sy = 1 if y1 < y2 else -1

    #initialise points storage
    xpoints = []
    ypoints = []

    #values for loop
    x = x1
    y = y1

    while x != x2 or y != y2:
        xpoints.append(x)
        ypoints.append(y)

        dp2 = 2*dp

        #update dp and x
        if dp2 > -dy:
            dp -= dy
            x += sx

        #update dp and y
        if dp2 < dx:
            dp += dx
            y += sy

    #plot line
    ax = plt.axis()
    xpoints.append(x)
    ypoints.append(y)


    plt.scatter(xpoints,ypoints)

```

```python
def drawtriangle(x1, y1, x2, y2, x3, y3):

  plt.title("Draw Triangle", fontdict=None, loc='center', pad=None)
  plt.xlim(-100, 100)  #setting x-axis length from -100 to 100
  plt.ylim(-100, 100)  #setting y-axis length from -100 to 100
  drawline(x1,y1,x2,y2)#
  drawline(x2,y2,x3,y3)#-goes from first to second to third to first creating a triangle.
  drawline(x3,y3,x1,y1)#
  plt.show()
```

```python
def rotatepoint():
  print("")
  print("Rotate Point")
  fig = plt.figure()
  ax = fig.add_subplot(projection='3d')

  #set initial coords
  x = -10
  y = 0
  z = 10

  #rotate on the x axis
  theta = np.radians(-300)  #set the rotation amount, convert deg to rad.
  y = y*(np.cos(theta)) - z*(np.sin(theta))
  z = y*(np.sin(theta)) + z*(np.cos(theta))

  ax.scatter(x, y, z, s=5)
  print("x = ", x, " y = ", y, " z = ", z)

  #rotate on the y axis
  theta = np.radians(175)
  z = z*(np.cos(theta)) - x*(np.sin(theta))
  x = z*(np.sin(theta)) + x*(np.cos(theta))

  ax.scatter(x, y, z, s=5)
  print("x = ", x, " y = ", y, " z = ", z)

  #rotate on the z axis
  theta = np.radians(65)
  x = x*(np.cos(theta)) - y*(np.sin(theta))
  y = x*(np.sin(theta)) + y*(np.cos(theta))

  ax.scatter(x, y, z, s=5)
  print("x = ", x, " y = ", y, " z = ", z)

  plt.show()
```

```python
def mainmenu():
  while True:
    print("")
    print("Maths Assignment 2")
    print(" (1)Draw the traingle")
    print(" (2)Transform the point")
    print(" (3)Draw your own line")
    print(" (4)Draw your own triangle")
    print(" (5)Exit")

    choice = int(input("Choose a function: "))

    if choice == 1:
        drawtriangle(-1,-1,60,10,20,30)

    if choice == 2:
        rotatepoint()

    if choice == 3:
        coordinates = input("Enter coordinates of two points (x1, y1, x2, y2): ")
        x1, y1, x2, y2 = map(int, coordinates.split(','))
        drawline(x1, y1, x2, y2)
        plt.xlim(-100, 100)  #setting x-axis length from -100 to 100
        plt.ylim(-100, 100)  #setting y-axis length from -100 to 100
        plt.show()

    if choice == 4:
        coordinates = input("Enter coordinates of 3 points (x1, y1, x2, y2, x3, y3): ")
        x1, y1, x2, y2, x3, y3 = map(int, coordinates.split(','))
        drawtriangle(x1, y1, x2, y2, x3, y3)
        plt.xlim(-100, 100)  #setting x-axis length from -100 to 100
        plt.ylim(-100, 100)  #setting y-axis length from -100 to 100

    if choice == 5:
      break

mainmenu()
```
